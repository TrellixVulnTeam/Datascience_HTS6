```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
train_X=data[['Fare']]
print(type(train_X))
train_Y=data[['Survived']]
train_X.head()
train_X.count()
```

```python
train_Y.head()
train_Y.count()
print(type(train_X))
```

```python
model.fit(train_X,train_Y)
```

```python
predictions = model.predict(train_X)
```

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(train_Y,predictions)
print(confusion)
print('    true positives  || false positives')
print('    ---------------------------------')
print('    false negatives || true negatives')
sns.heatmap(confusion)
plt.title('Heat map of confusion matrix')
plt.show()
```
